## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from functools import reduce
from collections import Counter

class BPETokenizer:

    def __init__(self, vocab_size: int) -> None:
        """Implements byte-pair encoding algorithm for tokenizing texts. Merges
        byte pairs until the size of the encoder's vocabulary is equal to
        `vocab_size`.

        Args:
            vocab_size (int): The maximum size of vocabulary.
        """
        self.vocab_size = vocab_size
        self.rules: list[tuple[str, str]] = []

    def _apply(self, rule: tuple[str, str], tokens: list[str]) -> list[str]:
        """Applies the learned rule (t_1, t_2) to the tokens by merging
        all pairs (t_1, t_2) in the token list together.
        
        Args:
            rule (tuple[str, str]): Rule to be applied.
            tokens (list[str]): List of existing tokens.
            
        Returns:
            list[str]: List of tokens after applying the provided rule.
        """ 
        result = []
        merged = False
        for s, t in zip(tokens, tokens[1:] + [None]):
            if merged:
                merged = False
                continue
            if (s, t) == rule:
                result.append(s + t)
                merged = True
            else:
                result.append(s)
        return result

    def _initial_tokenize(self, text: str) -> list[str]:
        """Extracts bytes from text.
        
        Args:
            text (str): Text to be tokenized.    
        
        Returns:
            list[str]: Tokenized text.
        """
        return list(text)

    def train(self, corpus: List[str]) -> None:
        """Trains the Byte-Pair Encoder on the given corpus. Merges byte pairs until
        the size of the vocabulary is equal to the required one.
        
        Args:
            corpus (list[str]): Training corpus.
        """
        # Tokenize the corpus
        corpus_tokens: list[list[str]] = list(map(self._initial_tokenize, corpus))
        # Calculate the initial vocabulary
        vocab: set[str] = reduce(set.union, map(set, corpus_tokens))
        while len(vocab) < self.vocab_size:
            byte_pair_counter = Counter()
            for tokens in corpus_tokens:
                byte_pairs = zip(tokens[:-1], tokens[1:])
                byte_pair_counter.update(byte_pairs)
            # Stop training if there is not enough tokens to form even one byte-pair
            if len(byte_pair_counter) == 0:
                break
            # The new rule is the most common byte-pair in the corpus
            (new_rule, _count), = byte_pair_counter.most_common(n=1)
            self.rules.append(new_rule)
            new_token = reduce(str.__add__, new_rule)
            vocab.add(new_token)
            # Apply the new rule to the corpus
            corpus_tokens = [self._apply(new_rule, tokens) for tokens in corpus_tokens]

    def tokenize(self, text: str) -> List[str]:
        """Tokenizes the text using learned byte-pairs.
        
        Args:
            text (str): Text to be tokenized.
            
        Returns:
            list[str]: Tokenized text.
        """
        tokens = self._initial_tokenize(text)
        for rule in self.rules:
            tokens = self._apply(rule, tokens)
        return tokens
```

```python
def test_BPETokenizer():
    """Tests BPETokenizer.
    """
    # Very small corpus, very big vocabulary size
    tokenizer = BPETokenizer(10**18)
    tokenizer.train(['ab', 'bc'])
    assert set(tokenizer.rules) == {('a', 'b'), ('b', 'c')}

    # Most common pair is added first
    tokenizer = BPETokenizer(4)
    tokenizer.train(['abab', 'abab'])
    assert tokenizer.rules == [('a', 'b'), ('ab', 'ab')]

    # Tokenization is performed correctly
    tokenizer = BPETokenizer(4)
    tokenizer.train(['abab', 'abab'])
    assert tokenizer.tokenize('abab') == ['abab']

    # Tokenizer can handle unseen tokens
    tokenizer = BPETokenizer(4)
    tokenizer.train(['abab', 'abab'])
    assert tokenizer.tokenize('ababc') == ['abab', 'c']

test_BPETokenizer()
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import glob
from pathlib import Path

DATA_PATH = Path('./data/')
assert DATA_PATH.exists()
assert DATA_PATH.is_dir()
```

```python
corpus = []
for file_path in glob.glob(str(DATA_PATH / '*')):
    with open(file_path, encoding='utf-8') as file:
        corpus += list(map(str.strip, file.readlines()))
```

```python
import numpy as np
import tqdm.notebook as tqdm

tweet_base_vocab_size = len(reduce(set.union, map(set, corpus)))
tweet_vocab_sizes = [tweet_base_vocab_size, 250, 500, 750, 1000]
tweet_vocab_size_to_average_tweet_len = {}
for vocab_size in tqdm.tqdm(tweet_vocab_sizes):
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(corpus)
    tokenized_corpus = map(tokenizer.tokenize, corpus)
    average_tweet_len = np.mean(list(map(len, tokenized_corpus)))
    tweet_vocab_size_to_average_tweet_len[vocab_size] = average_tweet_len
```

```python
import matplotlib.pyplot as plt
tweet_avg_lengths = [tweet_vocab_size_to_average_tweet_len[vocab_size] for vocab_size in tweet_vocab_sizes]
plt.plot(tweet_vocab_sizes, tweet_avg_lengths)
plt.xlabel('Vocab size')
plt.ylabel('Average tweet length')
plt.show()
```

### Additional plots with added vocabulary sizes

```python
import time

tweet_base_vocab_size = len(reduce(set.union, map(set, corpus)))
tweet_vocab_sizes = [tweet_base_vocab_size] + list(range(200, 1000 + 1, 50))
tweet_vocab_size_to_average_tweet_len = {}
tweet_vocab_size_to_training_time = {}
for vocab_size in tqdm.tqdm(tweet_vocab_sizes):
    tokenizer = BPETokenizer(vocab_size)
    train_start = time.perf_counter()
    tokenizer.train(corpus)
    train_end = time.perf_counter()
    tokenized_corpus = map(tokenizer.tokenize, corpus)
    average_tweet_len = np.mean(list(map(len, tokenized_corpus)))
    tweet_vocab_size_to_average_tweet_len[vocab_size] = average_tweet_len
    tweet_vocab_size_to_training_time[vocab_size] = train_end - train_start
tweet_avg_lengths = [tweet_vocab_size_to_average_tweet_len[vocab_size] for vocab_size in tweet_vocab_sizes]
tweet_train_times = [tweet_vocab_size_to_training_time[vocab_size] for vocab_size in tweet_vocab_sizes]
plt.plot(tweet_vocab_sizes, tweet_avg_lengths)
plt.xlabel('Vocab size')
plt.ylabel('Average tweet length')
plt.show()
plt.plot(tweet_vocab_sizes, tweet_train_times)
plt.xlabel('Vocab size')
plt.ylabel('Training time (seconds)')
plt.show()
```

### Task 2.3. Analysis on Dataset of Different Language (10 points)

Find a small dataset of texts in a language other than English. The dataset size should be not greater than several megabytes.

Train the BPE tokenizer on the dataset that you found. Try to tokenize the sentences from this dataset with the tokenizer of different `vocab_size`. Plot the dependency of the average length of the tokenized sentence by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized sentence on average.

Tell how how the average length of the tokenized sentence differs from the average length of the tokenized tweet. Explain why. 

```python
with open('war_and_peace.txt', encoding='utf-8') as file:
    sentences = [line.strip() for line in file.readlines()]
```

```python
import time

sentence_base_vocab_size = len(reduce(set.union, map(set, sentences)))
sentence_vocab_sizes = [sentence_base_vocab_size] + list(range(200, 500 + 1, 50))
vocab_size_to_average_sentence_len = {}
sentence_vocab_size_to_training_time = {}
for vocab_size in tqdm.tqdm(sentence_vocab_sizes):
    tokenizer = BPETokenizer(vocab_size)
    train_start = time.perf_counter()
    tokenizer.train(sentences)
    train_end = time.perf_counter()
    tokenized_sentences = map(tokenizer.tokenize, sentences)
    average_sentence_len = np.mean(list(map(len, tokenized_sentences)))
    vocab_size_to_average_sentence_len[vocab_size] = average_sentence_len
    sentence_vocab_size_to_training_time[vocab_size] = train_end - train_start
sentence_avg_lengths = [vocab_size_to_average_sentence_len[vocab_size] for vocab_size in sentence_vocab_sizes]
sentence_train_times = [sentence_vocab_size_to_training_time[vocab_size] for vocab_size in sentence_vocab_sizes]
plt.plot(sentence_vocab_sizes, sentence_avg_lengths)
plt.xlabel('Vocab size')
plt.ylabel('Average sentence length')
plt.show()
plt.plot(sentence_vocab_sizes, sentence_train_times)
plt.xlabel('Vocab size')
plt.ylabel('Training time (seconds)')
plt.show()
```
